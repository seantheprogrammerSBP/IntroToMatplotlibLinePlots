```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
```

```python
df.head()
```

```python
df.set_index('Country', inplace=True)
```

```python
df.head(3)
```

```python
# remove the name of the index
df.index.name = None
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print('Matplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
haiti = df.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line', title='Immigration from Haiti', xlabel='Years', ylabel='Number of Immigrants')
```

```python
haiti.index = haiti.index.map(int) 
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show() 
```

# Get the data set for China and India, and display the dataframe

```python
df_china_india = df.loc[['India', 'China'], years]
df_china_india
```

```python
df_china_india.plot(kind='line')
```

```python
df_china_india = df_china_india.transpose()
df_china_india.head()
```

```python
df_china_india.index = df_china_india.index.map(int)
df_china_india.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
inplace = True # paramemter saves the changes to the original df_can dataframe
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# get the top 5 entries
df_top5 = df.head(5)

# transpose the dataframe
df_top5 = df_top5[years].transpose() 

 #Step 2: Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(15, 8)) 

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
